```python
df = df_dynamic.dropna()
```

```python
# change dataframe to numpy array
df = df.values

# new number of rows and columns
n_rows, n_cols = df.shape
```

```python
df = df[np.lexsort((df[:, 0], df[:, 1]))] # sort by MMSI, then by time/date
```

```python
test = []
temp = []
mmsi = df[:,1]
unique_vals, unique_count = np.unique(mmsi, return_counts=True)
k = 0
last_count = 0
for i in range(len(df)):
    temp.append(tuple([df[i][3], df[i][2]]))
    if (i - last_count) == unique_count[k]:
        test.append(temp)
        temp = []
        last_count = i
        k += 1
test.append(temp)
```

# Folium可视化

```python
m = folium.Map(tiles="OpenStreetMap", zoom_start=6)
```

```python
for i in range(len(test[-300:])):
    for j in range(len(test[i])):
        folium.Circle(
            radius=2,
            location = test[i][j]
        ).add_to(m)
```

```python
for i in range(len(test[300:500])):
    for j in range(len(test[i])):
        folium.Circle(
            radius=2,
            location = test[i][j]
        ).add_to(m)
```

```python
m.save("aohai_20230130_mmsi_.html")
```
